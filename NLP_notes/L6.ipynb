# Generative AI

## Training Generative LLMs

### What is Generative LLM
- Generative LLMs are deep learning models that produces text
- Goals: 
    - AI systems trained to predict the next word given preceding text
- Training: 
    - typically fine-tuned to follow human instructions and generate responses aligned with human preferences
- Architecture: 
    - based on deep neural networks with billions of parameters
    - built on transformer models (with attention mechanisms, which endogenously assign varying degrees of importance to different words)

### Training Process
- Step 1: Pre-Training
    - This step is Self-Supervised Learning: model is feed Terabytes of text fragments (Wikipedia, scientific articles, books, etc.)
    - Next Token Prediction: the model predicts conditional probability distribution over words given the preceding words
    - Parameters are adjusted through back propagation
    - Through this step, neural nets learn language structure:
        - Syntactic structures
        - Relationships between words and concepts they represent
        - Context of sentences and word interactions
        - Relationships between different sentences
- Step 2: Instruction Fine-Tuning
    - This step is Supervised learning: feeding the model millions of examples for responding to thousands of different instructions
    - This instruction fine-tuning process improves model to follow human instructions
    - Example: before this step, the pretrained model might predict “And how old are you?” as continuation of “What is your name?”
    - This is not what we want, so we fintunes the model to respond appropriately
- Step 3: Reinforcement Learning
    - This step is Reinforcement Learning: improves model by incorporating human feedback
    - Human raters provide feedback on different responses
    - Helps align model with human preferences, particularly in areas hard to define via instruction fine-tuning (e.g. penalizing hateful responses)
    - However, this is also anosiy process, might explains why LLms become authoritative even when hallucinating:
        - Human prefers deterministic answer reather than vague answer
        - This promotes LLMs to give confident response even when they do not know the real answer

## Autoencoding Language Model

### What is Autoencoding Language Model
- They are transformer models pretrained by masking some input tokens and trying to reconstruct the original sequence
- Usually build bi-directional representations and get access to the full sequence 
- The model task: given both the sequences of words prior and after the masked word, predict the masked word
- The pre-trained model can be then fine-tuned and achieve great results on many tasks: e.g. text classification, sentiment analysis
- Main example of this class of models: BERT

### BERT
- BERT refers to Bidirectional Encoder Representations from Transformers
- Task: Masked language modeling
    - 15% of words selected in the training corpus is masked
    - If a word is masked: replace with [MASK] 80% of the time, a random token 10% of the time, and left unchanged 10% of the time
    - Not 100% replaced by [MASK] to prevent the model relying on [MASK] as a predictive feature (which does not exist in downstream tasks)
    - The model predicts the selected token using bidirectional context (it can observe to both left and right tokens via full self-attention)
- Architecture:
    - A stack of transformer blocks, each with a self-attention layer and an MLP
    - BERT-Large has 24 blocks, embedding dimension of 1024, and have 340M parameters in total to learn
- Corpus:
    - 800M words from English books (modern work, from unpublished authors)
    - 2.5B words of text from English Wikipedia articles
- Extensions: RoBERTa (Robust BERT), VisualBERT

## Autoregressive Language Model

### What is Autoregressive Language Model
- They are transformer models pretrained on classic language modeling task: guess the next token having read all the previous ones
- Distinction from autoencoding language model: during training, attention heads only view previous tokens, not subsequent tokens
- These models are ideal for text generation
- Main example of this class: GPT

### GPT
- GPT refers to Generative Pre-Trained Transformer
- GPT 1 (2018):
    - The first autoregressive transformer model
    - Trained on the Books corpus (7000+ books)
    - Pre-trained on a classical language modeling task, fintuined on supervised learning tasks (text classification, natural language inference, question answering)
- GPT 2 (2019):
    - Trained on all articles linked from Reddit with at least 3 upvotes (8 million documents, 40 GB of text)
    - Remove the supervised learning task, make some other architectural adjustments
    - Make model much bigger
- GPT 3 (2020):
    - Trained on an even bigger corpus (Common Crawl, WebText2, Books1, Books2 and Wikipedia)
    - Make model much much bigger
- GPT 3.5 (2022):
    - Subclass of GPT-3 trained on data up to June 2021
    - Incorporates the base model on which GPT3 is fine-tuned + is optimized for chat
- GPT 4 (2023):
    - Multimodal model: can take also images as inputs
    - Trained in two stages:
        - Next token prediction (like other GPT models)
        - Reinforcement learning with human feedback
        - Make model much much much bigger
- GPT 5.2 (2025):
    - Positioning: OpenAI flagship model for coding and agentic (tool-using) tasks
    - Context size: 400k context window
    - Output size: 
        - Up to 128k max output tokens (API)
        - Text + image input or pure text output
    - Reasoning control: based on reasoning.effort parameter
    - Knowledge cutoff: trained on data before Aug 31, 2025 (but can access external sources through RAG)

## Using Generative AI for Research

### Generative AI for Economic Research
- Main takeaway:
    - Employ AI agents in research is to treat them like a PI would treat a team of research assistants
    - They require careful planning of what is to be done, oversight during execution, and detailed vetting of the final results
- Good practices:
    - Provide context (e.g., specify point of view, audience, etc.)
    - Iterate
    - Be patinet

### Chatbots, Reasoners, and Agents
- Chabots: they are traditional LLMs (broadly available since Nov 2022), providing fast, fluent, single-pass responses
- Reasoning models: provide step-by-step analysis for harder math/logic/coding question (introduced Sep 2024)
- Agentic chatbots: combine language + reasoning + autonomous use of external tools (emerged Dec 2024) 

### What is an AI Agent
- An AI agent combines:
    - A large language model (LLM), also called foundation model
    - A planning process (decide steps)
    - Memory (track intermediate results after each step)
    - Autonomous tool access (web search, code execution, APIs, files)
- Unlike a standard chatbot, an AI agent can:
    - Decompose a question into subtasks
    - Gather information from multiple sources
    - Execute code and iterate based on intermediate findings
    - Synthesize results into coherent narratives
- The basic agent loop (key design idea is to make intermediate outputs explicit so they can be checked, logged, and debugged):
    1. Think: decide what information/tool calls are needed
    2. Act: call tools/APIs (e.g., data retrieval)
    3. Observe: parse/validate returned outputs
    4. Respond: produce a grounded natural-language answer

### Deep Research Agent for Literature Review
- Core architecture:
    - Lead/orchestrator plans the review process and decomposes the question into subtasks
    - Specialized subagents investigate (find and read papers) in parallel (each with its own context/tools)
    - Synthesis integrates findings into a cited report
- Strength: 
    - Compresses time-intensive information gathering + synthesis dramatically
- Limitation:
    - Can struggle to identify the most impactful frontier papers
    - Steering depth/breadth is constrained by provider cost limits and usage caps

### Limitations of AI Agents for Research
- Hallucinations (incl. bogus citations/statistics)
- Error propagation / multi-step cascades
- Brittleness to prompt changes (reproducibility issues)
- Prompt injection risks when tools/web are involved
- Not yet reliable as independent economic reasoners
