## Introduction

### Human Vision

### Computer Vision
- Computer vision: the computational process of interpreting and understanding visual data to derive information about the structure and properties of the external world
- Techniques of computer vision:
    - Classic techniques: pre-deep learning relied on handcrafted feature engineering, like color histograms and image entropy
    - Deep learning techniques: rely on neural networks for auctomatic feature learning, include following types
        - Convolutional neural networks: specialized neural networks for CV tasks, such as object detection, facial recognition, image segmentation, and sentiment analysis
        - Autotaggers: commerical model that allow us to skip model training
        - GANs: for image generation
        - Vision transformers: for obscured image classification

### What is an Image
- An image is a two-dimensional function $f(x, y)$, that maps a coordinate pair (a pixel coordinate) to an integer or vector related to intensity / color
- Single-channel image: 
    - The function maps a coordinate pair to an interger that represents grayscale value
    - The image can be represented as an array with width X and height Y
- Multiple-channel image: 
    - The function maps a coordinate pair to a vector
    - In RGB case, the vector at each pixel has three values, $(r_{x,y} , g_{x,y}, b_{x,y})$, representing channel-specific pixel value for red, green, and blue channels
    - Channel-specific pixel value can be an integer between 0 and 255 or a floating-point value in $[0, 1]$
    - The image can be represented as an array with width X, height Y, and depth 3

## Classical Machine Learning

### Pipeline in Classical Machine Learning
- Data collection and pre-processing:
    - Gather a labeled dataset of images 
    - Pre-process the images (e.g., resize, convert to grayscale, normalize).
- Feature extraction and selection:
    - Extract features that we think is relevant to the prediction task
    - The features "manually" crafted cannot be physical characteristics: such as color of eye, style of hair, etc.
    - It is impossible to detect these physical characteristics by specfying a general rule / using classical ML
    - Even if we do that, such a rule will not be robust to factors such as camera angle
    - Hence, features that we can extract must be general across images: such as color histogram and image texture
- Model training:
    - Train a model with the selected features
    - Note that we do not use the whole image, which is an array of pixel values, as the input of model
    - Instaed, we only use the manually-extracted features as input for the model
    - This is the biggest distinction from deep learning CV techniques
- Model evaluation:
    - Use metrics and cross-validation to assess model performance
- Model deployment and prediction:
    - Deploy the trained model for making predictions on new images

### Feature Extraction in Classical ML
- Different feature types provide a different kind of information about the image content
- The best one to use depends on the specific task at hand and rely on domain knowledge
- Below, we outlisted some common features that can be extracted from images and used for classical ML:
    - Color Histograms: represents the distribution of color values in an image
    - Texture Features: calculate texture features like entropy, energy, contrast, and homogeneity from the grayscale co-occurrence matrix
    - Haar-like Features: use integral images to efficiently compute features that recognize edges, lines, and rectangles for object detection

### Color Histogram
- Color histogram for a channel represents the distribution of channel-specific pixel value (intensity levels)
- Mathematically, the color histogram is the function $h(r_k) = n_k$
    - where $r_k$ is the k-th intensity value
    - $n_k$ is the number of pixels with intensity $r_k$
- Typical usage of this feature: for image retrieval and color-based segmentation tasks

### Texture Feature
- Texture features are typically based on the grayscale co-occurrence matrix P:
    - This matrix contains the occurrences of pairs of grayscale values at a given offset (distance and direction)
    - For example, we can specify the offset as 1 pixel to the right
    - Then, P counts the occurence of pixel value j at 1 pixel right to pixel value i for every possible (i, j)
    - Normalize this matrix gives us empirical P(i, j) for every possible (i, j)
- One common texture feature is texture entropy:
    - Entropy measures the amount of information or randomness in the image region
    - The texture entrophy of an image can be defined as $E = - \sum_{i,j} P(i,j) \log \big( P(i,j) \big)$, where $P(i, j)$ comes from P
    - Higher entropy indicates a more complex texture with more details in the image
    - Lower entropy indicates a more uniform texture with fewer details
    - Despite texture entrophy, we can also calculate local entrophy:
        - Texture entrophy is at image level, local entrophy is at pixel level
        - For each pixel, take a small window around it
        - Count occurence of gray-scale values in that window and normalize to probabilities
        - Compute entropy

### Haar-like Features
- Haar-like features represent differences in intensity between rectangular regions of an image
- Firstly, define a set of rectangular regions with different positions and size, in order to represent face brightness contrast patterns
- Then calculate the feature by summing the pixel intensities in specific regions and subtracting one region’s sum from another
- Helpful for face detection: it is a structured intensity comparison between neighboring rectangular regions that captures face contrast patterns

## Deep Learning

### Classical ML vs. Deep Learning
- Classical ML:
    - Manual Feature Engineering: requires domain expertise to design and extract features like edges, textures, shapes, or color histograms from images
    - Do not and cannot take the whole image (array of pixel values) as model input
    - Feature extraction is a distinct (and manual) step performed before training the model
    - The effectiveness of the model depends heavily on the quality and relevance of manually extracted features
- Deep Learning:
    - Automatic Feature Learning: pixels are basic features, take whole image as input, find abstarct feature through hidden layers
    - Integrated Process: features are learned directly from the data during training
    - Hierarchical Features: learns low-level features (edges, corners) in early layers and high-level abstract features (object shapes, patterns) in deeper layers
