# Unsupervised Learning with Text

## Road Map

### Different Methods for Different Goals
- Unsupervised Learning:
    - Algorithm discovers patterns in text (or other high-dimensional data)
    - Human interprets and explain the results (e.g. interpret the content of topics or clusters)
- Supervised Learning:
    - Predicting a known lablel, e.g., predicting whether a political speech is from a Democrat or a Republican
    - Machine learns to replicate labels for new data points
- Both strategies amplify human effort, each in different ways:
    - Unsupervised learning need human to interpret the pattern
    - Supervised learning need human to label the training data
- Distinctions between two methods in research are not clear-cut:
    - Unsupervised learning models' outputs can be used as input/feature for supervised learning model
    - Supervised learning models can be used to discover themes/patterns

### Steps in Social-Science Research using Unsupervised Learning
1. What is the research question?
2. Get Corpus and Data:
    - obtain, clean, preprocess, and link text with metdata
    - produce descriptive visuals and statistics on the text and metadata
3. Unsupervised learning:
    - What pattern are we trying to measure?
    - Select a model and train it
    - Prove model robustness to hyperparameters
    - Validate that the model is measuring what we want (interpret the pattern + validate it using metdata)
4. Empirical analysis:
    - Produce statistics or predictions with the patterns found by trained model
    - Answer the research question

## Dimensionality Reduction

### The Basics of Dimensionality Reduction
- Why dimensionality reduction: 
    - Word matrix built using a lareg vocabulary is very high-dimensional
    - Document vectors have large amount of elements (each document represented in the high dimensional feature space)
    - Directly use the document vector as input for prediction lead to curse of dimensionality (esaily overfit)
    - Therefore, there is a need to conduct dimensionality reduction on document vectors
- Assumption:
    - Documents are not distributed uniformly across the feature space (frequency of each words)
    - They have a lower-dimensional latent structure – a manifold – that can be learned
    - Dimensionality reduction makes data more interpretable – for example by projecting down to two dimensions for visualization
    - It may also help improves computational tractability / model performance

### Principal Component Analysis
- Principal component analysis (PCA) is the basic method for dimension reduction
- PCA computes the dimension in data explaining most variance as the first component
- After the first component, subsequent components learn the (orthogonal) dimensions explaining most variance in dataset after projecting out first component
- The number of components kept is the hyperparameter of PCA, it is usually set through elbow appraoch:
    - Compute explained variance of the data by component number
    - Find the component number such that there is no laonger sharp increase in explained variance

### PCA and LSA
- The document-term matrix X can be reduced by projecting down to first principal component dimension using PCA
- This is known as latent semantic analysis (LSA)
- Advantage of LSA:
    - Distance metrics between documents (e.g. cosine similarity) are approximately preserved
- Limitation of PCA and LSA:
    - PCA factors (principal dimension) are not interpretable
    - For non-negative data (e.g. counts or frequencies), Non-negative Matrix Factorization (NMF) provides more interpretable factors than PCA

## Topic Models 

### Topic Models in Economics
- Core methods for topic models were developed in computer science and statistics:
    - They summarize unstructured text by topic distribution
    - They use words within document to infer topic
    - They are useful for dimension reduction
- Economists apply these method and use inferred topic distribution as a form of measurement:
    - How observed covariates drive trends of topics in language
    - Tell a story not just about what (descriptive trend), but how and why (what covariate drives the distribution of topics)
    - Topic models are more interpretable than other dimension reduction methods, such as PCA

### Latent Dirichlet Allocation
- Assumption:
    - Each topic is a distribution over words
    - Each document is a distribution over topics
    - There are K topics (tunable hyperparameter)
- Input: N × M document-term count matrix X
- Output: Estimates of probabilities that each word and each document are composed of each topic
- Like NMF, LDA works by factorizing X into:
    - An N × K document-topic matrix
    - An K × M topic-term matrix
- Unlike NMF, LDA is probabilistic:
    - LDA is a generative probabilistic model, meaning it assumes documents were generated by an underlying topic structure, and it tries to reverse-engineer that structure
    - It achieves so by estimating the latent (hidden) topic structure that best explains the word co-occurrences in documents
    - In practice, algorithm maximizes the likelihood of observed documents given the latent topic structure
- Hyper-parameter:
    - K: number of topics, can be selected using coherence metrics
    - Dirichlet priors: 
        - Alpha controls topics per document: small alpha, small number of topics per document, each with high probability
        - Beta controls words per topic: small beta, small number of words per topic, each with high probability
- Usage:
    - Once trained, can easily get topic proportions for a corpus, can then use the topic proportions as variables in a economics analysis
    - Use the trained model to predict topic disstribution for new document (not in training sample)
    - Main topic is the highest-probability topic in a document
    - Documents with highest share in a topic work as representative documents for the topic

### Structural Topic Model
- Structural Topic Model, also called STM, is a combination of LDA with meta-data
- STM incoportaes metdata through two ways:
    - Topic distribution can vary by metadata (e.g. Republicans talk about military issues more then Democrats)
    - Topic content can vary by metadata (e.g. Republicans talk about military issues more patriotically than Democrats)
- Structural topic model is not a prediction model:
    - It will tell which topics or features correlate with an metdata outcome, but it will not provide an in-sample or out-of-sample prediction for an outcome
    - It uses another distribution of the priors (not Dirichlet) such that without metadata it replicates the correlated topic model 

### Recent Advances in Topic Model
- Keyword-Assisted topic model (Eshima, Imai, and Sasaki, 2024)
    - Allows semi-supervised creation of topics
    - Input seed dictionaries and labeled topics
- Problems with unstructured data and causal inference (Battaglia et al., 2024)
    - Shows that two-steps strategy (first apply topic model to infer topic distribution, then use distribution as a covariate) leads to invalid inference
    - Propose solutions with bias correction and a one-step strategy

## Supervised Machine Learning with Text

### A baseline for supervised machine learning using text
1. Take POS-filtered bigrams as inputs X
2. Select a machine learning model for predicting outcome y:
    - For regression (y is one-dimensional and continuous), elastic net or gradient boosted regressor
    - For classification (y is discrete), L2-penalized logistic regression or gradient boosted classifier
3. Use cross-validation grid search in training set to select model hyperparameters
    - For classification, use cross entropy
    - For regression, use mean squared error as loss function
4. Evaluate model in held-out test set:
    - For classification, use balanced accuracy, confusion matrix, and calibration plot (more in slide)
    - For regression, use R squared and binscatter plot
5. Interpret the model predictions:
    - For linear models, examine coefficients
    - For random forest/gradient boosting, use feature importance ranking
    - Look at highest and lowest ranked feature for predicting y
6. Answer the research question
